```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
from glob import glob
from pyphot import (unit, Filter)

telescope = os.listdir("filters")
telescope = [t for t in telescope if t[0]!='.']


instruments = []
for t in telescope:
    filters = os.listdir("filters/"+t)
    instrument = [f"{t}/{f.split('.')[0]}" for f in filters]
    instrument = list(dict.fromkeys(instrument))
    instruments.extend(instrument)

indexing = []
for instru in instruments:
    filters = glob(f"filters/{instru}*")
    # print(filters)
    wavelength = []
    for fil in filters:
        print(fil)
        df = pd.read_csv(fil, delim_whitespace=True, header=None, names=['wavelength', 'transmission'], comment='#')
        wavelength.append(Filter(df['wavelength'], df['transmission'], name='xxxxx', dtype='photon', unit='Angstrom').lpivot.value)
    # dddf['filters'] = filters
    ddff = pd.DataFrame({"w":wavelength, "f":filters})
    ddff.sort_values(by=['w'], inplace=True)
    filters = ddff['f'].values

    filters = [".".join(f.split('/')[-1].split('.')[1:]) for f in filters]
    indexing.append({
        "value":instru,
        "label":instru,
        "dialog":True,
        "children":[{"value":instru+"."+fil, "label":fil[:-4], "checked":False} for fil in filters]})
indexing
```

```python
# Directly from dictionary
with open('./myfilter_frontend/filters/indexing.json', 'w') as outfile:
    json.dump(indexing, outfile)
```
